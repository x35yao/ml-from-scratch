### Sanity check with synthetic function

```python
%load_ext autoreload
%autoreload 2

import numpy as np
from mlscratch.utils import train_test_split
from mlscratch.random_forest import RandomForestRegressor

x = np.linspace(-3, 3, 200)
y = np.sin(x)

X_train, y_train, X_test, y_test = train_test_split(x, y)
model = RandomForestRegressor(n_estimators = 5)
model.fit(X_train, y_train)
y_hat = model.predict(X_test)

from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(x, y, '--', color = 'r', label = 'Ground truth function')
ax.scatter(X_test, y_hat, label = 'Test points')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```
